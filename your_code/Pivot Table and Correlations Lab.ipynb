### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
def remove_maxima_calculate_mean(group):
    group = group[group['Salary'] < group['Salary'].max()]
    return group['Salary'].mean()

mean_salary_after_removing_maxima = employee.groupby(['Department'])[['Salary']].apply(remove_maxima_calculate_mean)
```
